```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

plt.style.use('fivethirtyeight')
elec_car= pd.read_csv('전기자동차_최종.csv', encoding = 'utf-8')
norm_car = pd.read_csv('일반자동차_최종.csv', encoding = 'utf-8')
elec_car
```

```python
#서울 행 추출
data_s_1 = elec_car.loc[0] #elec_car 데이터프레임의 0번째 행 추출
ndata_s_1 = norm_car.loc[0] #norm_car 데이터프레임의 0번째 행 추출
#print(data_s) 
#위의 열도 같이 출력된다
data_s_1 = data_s_1.values.reshape(1,-1) #데이터 한줄로 만들기
data_s_1 = data_s_1.tolist() #리스트로 보내기
data_s_1 = sum(data_s_1, [])
data_s_1 = data_s_1[1:]
#print(data_s)
ndata_s_1 = ndata_s_1.values.reshape(1,-1) #데이터 한줄로 만들기
ndata_s_1 = ndata_s_1.tolist() #리스트로 보내기
ndata_s_1 = sum(ndata_s_1, [])
ndata_s_1 = ndata_s_1[1:]
```

```python
#인천 행 추출
data_s_3 = elec_car.loc[2] #elec_car 데이터프레임의 2번째 행 추출
ndata_s_3 = norm_car.loc[2] #norm_car 데이터프레임의 2번째 행 추출
#print(data_s) 
#위의 열도 같이 출력된다
data_s_3 = data_s_3.values.reshape(1,-1) #데이터 한줄로 만들기
data_s_3 = data_s_3.tolist() #리스트로 보내기
data_s_3 = sum(data_s_3, [])
data_s_3 = data_s_3[1:]
#print(data_s)
ndata_s_3 = ndata_s_3.values.reshape(1,-1) #데이터 한줄로 만들기
ndata_s_3 = ndata_s_3.tolist() #리스트로 보내기
ndata_s_3 = sum(ndata_s_3, [])
ndata_s_3 = ndata_s_3[1:]
```

```python
#대구 행 추출
data_s_4 = elec_car.loc[3] #elec_car 데이터프레임의 3번째 행 추출
ndata_s_4 = norm_car.loc[3] #norm_car 데이터프레임의 3번째 행 추출
#print(data_s) 
#위의 열도 같이 출력된다
data_s_4 = data_s_4.values.reshape(1,-1) #데이터 한줄로 만들기
data_s_4 = data_s_4.tolist() #리스트로 보내기
data_s_4 = sum(data_s_4, [])
data_s_4 = data_s_4[1:]
#print(data_s)
ndata_s_4 = ndata_s_4.values.reshape(1,-1) #데이터 한줄로 만들기
ndata_s_4 = ndata_s_4.tolist() #리스트로 보내기
ndata_s_4 = sum(ndata_s_4, [])
ndata_s_4 = ndata_s_4[1:]
```

```python
#제주 행 추출
data_s_5 = elec_car.loc[4] #elec_car 데이터프레임의 4번째 행 추출
ndata_s_5 = norm_car.loc[4] #norm_car 데이터프레임의 4번째 행 추출
#print(data_s) 
#위의 열도 같이 출력된다
data_s_5 = data_s_5.values.reshape(1,-1) #데이터 한줄로 만들기
data_s_5 = data_s_5.tolist() #리스트로 보내기
data_s_5 = sum(data_s_5, [])
data_s_5 = data_s_5[1:]
#print(data_s)
ndata_s_5 = ndata_s_5.values.reshape(1,-1) #데이터 한줄로 만들기
ndata_s_5 = ndata_s_5.tolist() #리스트로 보내기
ndata_s_5 = sum(ndata_s_5, [])
ndata_s_5 = ndata_s_5[1:]
line=[]
data_i = elec_car.iloc[:0,1:] #년 월 시리즈 추출
for row in data_i:
    line.append(row[:])
```

```python
percentage_s_1 = [] #서울 퍼센테이지
for i in range(36):
    print(round((data_s_1[i]/ndata_s_1[i])*100,3))
    
    percentage_s_1.append(round((data_s_1[i]/ndata_s_1[i])*100,3))
#percentage_s.append(data_s[i] / ndata_s[i])
#print(percentage_s)
```

```python
percentage_s_2 = [] #경기 퍼센테이지
for i in range(36):
    print(round((data_s_2[i]/ndata_s_2[i])*100,3))
    
    percentage_s_2.append(round((data_s_2[i]/ndata_s_2[i])*100,3))
```

```python
percentage_s_3 = [] #인천 퍼센테이지
for i in range(36):
    print(round((data_s_3[i]/ndata_s_3[i])*100,3))
    
    percentage_s_3.append(round((data_s_3[i]/ndata_s_3[i])*100,3))
```

```python
percentage_s_4 = [] #대구 퍼센테이지
for i in range(36):
    print(round((data_s_4[i]/ndata_s_4[i])*100,3))
    
    percentage_s_4.append(round((data_s_4[i]/ndata_s_4[i])*100,3))
```

```python
plt.rc('font', family='NanumGothic')

fig = plt.figure(figsize=(15,15))
fig.set_facecolor('white')
ax = fig.add_subplot(1,1,1)

xticks = ax.get_xticks()
ax.plot(line,data_s_1,marker='',label='서울', color = 'r')
ax.plot(line,data_s_2,marker='',label='경기', color = 'blue')
ax.plot(line,data_s_3,marker='',label='인천', color = 'green')
ax.plot(line,data_s_4,marker='',label='대구', color = 'skyblue')
ax.plot(line,data_s_5,marker='',label='제주', color = 'pink')


ax.set_xticks(line[2::3]) #2번째 열부터 모든 열 3칸씩 증가하기
ax.tick_params(labelsize =12 , length = 20, width =3)
#plt.xlabel('월', fontsize = '20' , loc = 'left')
#plt.xlabel('x axis', fontsize=14, labelpad=20.0)
plt.xticks()
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%.lf 대'))
ax.legend()
plt.title('월별 제주 전기자동차 수',fontsize=20)
plt.show()
```
